```python
# def for countplot
def count_plot(a,b,c):
    ax = sns.countplot(x=a,data =b,hue=c)
    for bars in ax.containers:
        ax.bar_label(bars)
```

```python
# def for barplot
def bar_plot(x,y,data,hue):
    sns.barplot(x=x,y=y,data=data,hue=hue)
```

### State

```python
# total number of orders from top 10 states
```

```python
sales_State_orders= df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
```

```python
bar_plot('State','Orders',sales_State_orders,None)
```

```python
# total amount/sales from top 10 states
sales_State_amount= df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
bar_plot('State','Amount',sales_State_amount,None)
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*
